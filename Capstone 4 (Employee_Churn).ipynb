# Predictive Analysis

```python
hr_churn.columns
```

```python
hr_churn=hr_churn.drop(columns=["Department"])
hr_churn.head()
```

Step 1

```python
X=hr_churn.drop(columns=["left"])
```

```python
X.head()
```

```python
y=hr_churn['left']
```

```python
y.head()
```

Rules

```python
X.dtypes
```

```python
X=pd.get_dummies(hr_churn,columns=["salary"],drop_first=True,prefix="DM")
```

```python
X.dtypes
```

```python
type(X)
```

```python
X.shape
```

```python
X.isna().sum()
```

```python
y.isna().sum()
```

Step 2

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42,stratify=y)
```

Step 3

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
forest=RandomForestClassifier()
```

```python
forest.fit(X_train,y_train)
```

Step 4

Evaluate

```python
forest.score(X_test,y_test)
```

Predict

```python
forest.predict(X_test[:1])                                          #Predict the output of 1st Row
```

```python
forest.predict(X_test)
```

```python
forest.predict(X_test[10:11])                                       #Predict the output of 10th Row 
```
